```python
def convert_data(dff):
    dic = {
        'darija': [],
        "eng": []
    }

    for index, row in dff.iterrows():
        row_list = [row["n1"], row["n2"], row["n3"], row["n4"]]

        for value in row_list:
            if type(value) != float:
                dic["darija"].append(value)
                dic["eng"].append(row["eng"])
    new_df = pd.DataFrame(dic)
    return new_df 
    
```

```python
def convert_verbs(dff):
    dic = {
        'darija': [],
        "eng": []
    }

    for index, row in dff.iterrows():
        row_list = [row["n1"], row["n2"], row["n3"], row["n4"],row["n5"],row["n6"]]

        for value in row_list:
            if type(value) != float:
                dic["darija"].append(value)
                dic["eng"].append(row["eng"])
    new_df = pd.DataFrame(dic)
    return new_df 
    
```
